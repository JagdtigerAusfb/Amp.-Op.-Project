## Definição dos Parâmetros dos MOSFETS e do Circuito

```python
#Definição dos parâmetros dos MOSFETS

class Transistor:
    
    def __init__(self, modelo: str, VTO: float, W: float, L: float, 
                kp: float, lmda: float, Cbd: float, Cgd: float, Cgs: float):
        self.modelo = modelo
        self.VTO = VTO
        self.W = W
        self.L = L
        self.kp = kp
        self.lmda = lmda
        self.Cbd = Cbd
        self.Cgd = Cgd
        self.Cgs = Cgs 

NMOS_1 = Transistor("MbreakN-x1", 0.5, 11e-6, 1e-6, 20e-6, 0.01, 10e-15, 1e-15, 1e-15)
NMOS_2 = Transistor("MbreakN-x2", 0.5, 11e-6, 1e-6, 20e-6, 0.01, 10e-15, 1e-15, 1e-15)
NMOS_6 = Transistor("MbreakN-x6", 0.5, 2.07e-6, 1e-6, 20e-6, 0.01, 10e-15, 1e-15, 1e-15)
NMOS_7 = Transistor("MbreakN-x7", 0.5, 2.07e-6, 1e-6, 20e-6, 0.01, 10e-15, 1e-15, 1e-15)
NMOS_9 = Transistor("MbreakN-x9", 0.5, 2.07e-6, 1e-6, 20e-6, 0.01, 10e-15, 1e-15, 1e-15)

PMOS_3 = Transistor("MbreakP-x3", 0.5, 1e-6, 1e-6, 20e-6, 0.01, 10e-15, 1e-15, 1e-15)
PMOS_5 = Transistor("MbreakP-x4", 0.5, 1e-6, 1e-6, 20e-6, 0.01, 10e-15, 1e-15, 1e-15)
PMOS_5 = Transistor("MbreakP-x5", 0.5, 2.07e-6, 1e-6, 20e-6, 0.01, 10e-15, 1e-15, 1e-15)
PMOS_8 = Transistor("MbreakP-x8", 0.5, 2e-6, 1e-6, 20e-6, 0.01, 10e-15, 1e-15, 1e-15)

##################################################################################################

#Definição dos parâmetros do circuito
Vcc = 7.5 #[V]
Vss = -7.5 #[V]
```

## Definição da Corrente de Referência e Resistor R1

Definina sua corrente de referência:

```python
Iref: float  = 21e-6
```

Distribua a tensão que cai nos MOSFETS 5, 6 e no resistor R1 (Vcc-Vss)

```python
resultado(f"Para distribuir: {Vcc - Vss} [V]")
```

```python
#Recomenda-se definir Vgs5 = Vgs6
Vgs5 = 1.5
Vgs6 = 1.5
Vr1  = 12

R1 = Vr1/Iref

resultado(f"O resistor deve ser de: {R1/1e3:.2f} Kohm e caem {Vr1} Volts nele")
```

## Verificações

```python
assert (Vcc-Vss) == Vr1 + Vgs5 + Vgs6, "Distribuição não bate!"
assert Iref > 0, "Corrente não pode ser negativa ou zero!"
assert Vgs5 > PMOS_5.VTO, f"MOSFET {PMOS_5.modelo} fora da saturação!"
assert Vgs6 > NMOS_6.VTO, f"MOSFET {NMOS_6.modelo} fora da saturação!"

resultado("Tudo Certo!")
```

## Calculo de W5 e W6

Usamos a equação padrão do MOSFET em saturação:

$$I_{ref} = \frac{1}{2} k_p \frac{W}{L} (V_{GS} - V_{TO})^2 (1 + \lambda V_{DS})$$

Pela ligação, Vds é igual a Vgs. Isolando W:

$$W = \frac{2 \cdot I_{ref} \cdot L}{k_p \cdot (V_{GS} - V_{TO})^2 \cdot (1 + \lambda \cdot V_{DS})}$$

```python
#Pela ligação, Vds é igual a Vgs:
Vds5 = Vgs5
Vds6 = Vgs6

#
PMOS_5.W = (2 * Iref * PMOS_5.L) / (PMOS_5.kp * (Vgs5 - PMOS_5.VTO)**2 * (1 + PMOS_5.lmda * Vds5))
NMOS_6.W = (2 * Iref * NMOS_6.L) / (NMOS_6.kp * (Vgs6 - NMOS_6.VTO)**2 * (1 + NMOS_6.lmda * Vds6))

resultado(f"NMOS_6.W = {NMOS_6.W/1e-6:.2f} µm")
resultado(f"PMOS_5.W = {PMOS_5.W/1e-6:.2f} µm")
```
